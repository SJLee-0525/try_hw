```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

# pandas   
netflix = pd.read_csv('archive/NFLX.csv')
# to_datetime          
netflix['Date'] = pd.to_datetime(netflix['Date'])
netflix_df_2022 = netflix[netflix['Date'] >= "2022-01-01"] 
netflix_df_2022 = netflix_df_2022.loc[:, ['Date', 'High', 'Low', 'Close']]
# netflix_df_2022

#    1
# netflix_df_2021_Mth = netflix_df_2021.groupby('Date', as_index = False).mean()

#    2 + groupby   average  mean
# netflix_df_2021_Mth = netflix_df_2022.groupby(pd.Grouper(key="Date", axis=1, freq="M")).mean().reset_index()

netflix_df_2022

#  
plt.plot(netflix_df_2022['Date'], netflix_df_2022['High'], label = 'High')
plt.plot(netflix_df_2022['Date'], netflix_df_2022['Low'], label = 'Low')
plt.plot(netflix_df_2022['Date'], netflix_df_2022['Close'], label = 'Close')

# 
plt.legend()

#      
plt.title('NFLX Average Close Price')
plt.xlabel('Date')
plt.ylabel("Average Close Price")

# x  
plt.xticks(rotation = 45)

#  
plt.show()
```
